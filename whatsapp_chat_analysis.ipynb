```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with B SECTION UNOFFICIAL.txt', 'r' , encoding='utf-8')
```

```python
data = f.read()
```

```python
# Restore the built-in print function if it was overwritten
print((data))
```

```python
# Define the regex pattern for WhatsApp timestamp format
pattern = r'(\d{1,2}/\d{1,2}/\d{2,4}),\s(\d{1,2}:\d{2})\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
len(messages)
```

```python
dates = re.findall(pattern, data)
dates
```

```python
min_len = min(len(messages), len(dates))
messages = messages[:min_len]
dates = dates[:min_len]
dates
```

```python
# Combine date and time into single string
date_strings = [f"{date[0]}, {date[1]}" for date in dates]

df = pd.DataFrame({'user_message': messages, 'message_date': date_strings})

# Convert message_date type - use correct format
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M')
df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
df.shape
```

```python
users = []
messages = []

# df['user_message'] me pehle se chat lines hain
for message in df['user_message']:
    entry = re.split(r'([\w\W]+?):\s', message)
    
    if entry[1:]:   # agar user name mila
        users.append(entry[1])
        messages.append(entry[2])
    else:   # system notifications ya group msgs
        users.append('group_notification')
        messages.append(entry[0])

# naya dataframe columns add karte hain
df['user'] = users
df['message'] = messages

# ab user_message column hata dete hain
df.drop(columns=['user_message'], inplace=True)

df.head(100)
```
